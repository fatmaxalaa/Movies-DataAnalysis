## 3- Show the Comedy movies released in 2009?

## 6- show top 10 directors

```python
new_df['director'].value_counts().head(10)
```

```python
top_10Dir = new_df.nlargest(10 , 'runtime')[['name' , 'runtime' ,'director']].set_index('name')
```

```python
top_10Dir
```

```python
sns.barplot(x='runtime' , y=top_10Dir.index ,data = top_10Dir)
```

## 7- Show the movies that Will Smith was in it ?

## 8-Show the Comedy Movies that have the rating of 'PG-13' , after year 2017?
